```python
import pandas as pd
```

```python
import random
```

```python
# Listas para gerar dados variados
departamentos = ['TI', 'Vendas', 'Marketing', 'Financeiro', 'RH']
fontes = ['LinkedIn', 'Indicação', 'Site da Empresa', 'Job Board']
status_opcoes = ['Contratado', 'Reprovado']
```

```python
data = []
for i in range(1, 101):
    data.append({
        'ID': i,
        'Departamento': random.choice(departamentos),
        'Fonte': random.choice(fontes),
        'Entrevista_Nota': round(random.uniform(5.0, 10.0), 1),
        'Status': random.choice(status_opcoes),
        'Tempo_Dias': random.randint(3, 45)
    })
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('dados_recrutamento_100.csv', index=False)
```
